```python
tokens = ['@', 'http', 'https', ',', '.', ':', ';', '!', '\"', '?']
def RemoveTokens(text):
  text = word_tokenize(text)

  text = [word for word in text if word not in tokens]
  text = " ".join(text)
  text = emoji.replace_emoji(text, replace='')

  return text

cleaned_data = []
for data in X:
  data1 = RemoveTokens(data)
  cleaned_data.append(data1)

df['cleaned_data'] = cleaned_data
print(df)
```

```python
stop_data = []
stop_list = []
for data in cleaned_data:
  data1, stop = RemoveStopWords(data)
  stop_data.append(data1)
  stop_list.append(stop)

stop_data_series = pd.Series(stop_list)

tokenized_text = [word_tokenize(text) for text in stop_data_series]
all_words = [word for sublist in tokenized_text for word in sublist]
s = pd.Series(all_words)
cnt = s.value_counts()
top = cnt.head(15)
print('Stop Words')
print(top)

df['stop_data'] = stop_data
```

```python
plane_words = ['americanair', 'united', 'delta', 'southwestair', 'jetblue', 'virginamerica', 'usairways', 'flight', 'plane', 'AmericanAir', 'United', 'Delta', 'SouthwestAir', 'JetBlue', 'VirginAmerica', 'USAirways']
def RemovePlane(text):
  text = word_tokenize(text)
  plane = [word for word in text if word in plane_words]
  text = [word for word in text if word not in plane_words]
  text = " ".join(text)
  plane = " ".join(plane)

  return text, plane
```

```python
plane_data = []
plane_list = []
for data in stop_data:
  data1, plane = RemovePlane(data)
  plane_data.append(data1)
  plane_list.append(plane)

plane_data_series = pd.Series(plane_list)

tokenized_text = [word_tokenize(text) for text in plane_data_series]
all_words = [word for sublist in tokenized_text for word in sublist]
s = pd.Series(all_words)
cnt = s.value_counts()
top = cnt.head(15)
print('Plane Words')
print(top)

df['plane_data'] = plane_data
```
